```python
#Load the Datasets
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("House_price.csv")
```

```python
df.dtypes
```

```python
#Data visualization libraries
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Univarient visualization

#Scatter plot
plt.scatter(df.index,df['Price'])
plt.show()
```

```python
sns.scatterplot(x=df.index,y=df['Price'],hue=df['Built Year'])
```

```python
plt.figure(figsize=(6,6))
plt.title('Line plot of House Price')
plt.xlabel('index')
plt.ylabel('Price')
plt.plot(df.index,df['Price'],markevery=1,marker='d')
for name,group in df.groupby('Built Year'):
    plt.plot(group.index,group['Price'],label=name,marker='d')
plt.legend()
plt.show()
```

```python
sns.set(rc={'figure.figsize':(7,7)})
sns.set(font_scale=1.5)
fig=sns.lineplot(x=df.index,y=df['Price'],markevery=1,marker='d',data=df,hue=df['Built Year'])
fig.set(xlabel='index')
```

```python
#Histograms
plt.hist(df['Price'])
```

```python
#Density Plots
plt.figure(figsize=(5,5))
df['Price'].plot(kind='density')
```

```python
#Rug plots
fig,ax=plt.subplots()
sns.rugplot(df['Price'])
ax.set_xlim(3,9)
plt.show()
```

```python
#box Plot
plt.boxplot(df['Price'])
```

```python
dfm=df.drop('Built Year',axis=1)
```

```python
#violin plots
plt.figure(figsize=(7,7))
plt.violinplot(dfm.values,showmedians=True)
```

```python
#Bar Chart
df['Built Year'].value_counts().plot.bar()
```

```python
#Pie chart
plt.pie(df['Built Year'].value_counts(),shadow=True)
```

```python
#Bivariate Analysis
#Scatter Plot
plt.scatter(df['Price'],df['Built Year'])
plt.title('Price of House with built Year ')
plt.xlabel('Price')
plt.ylabel('Built Year')
```

```python
#Correlation coefficients
df.corr()
```

```python
#simple Linear Regression
import statsmodels.api as sm
#define response variable
y = df['Price']

#define explanatory variable
x = df[['Built Year']]

#add constant to predictor variables
x = sm.add_constant(x)

#fit linear regression model
model = sm.OLS(y, x).fit()

#view model summary
print(model.summary())
```
